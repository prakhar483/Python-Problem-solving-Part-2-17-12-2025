```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            num = start + 2 * i
            result ^= num   # bitwise XOR
        return result
```

```python
sol = Solution()

n = 5
start = 0
sol.xorOperation(n, start)
```

```python
sol.xorOperation(4, 3)
```

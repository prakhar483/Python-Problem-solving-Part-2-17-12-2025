```python
class Solution:
    def findNumbers(self, nums):
        count = 0
        
        for num in nums:
            # convert number to string and count digits
            if len(str(num)) % 2 == 0:
                count += 1
        
        return count
```

```python
sol = Solution()

nums = [12, 345, 2, 6, 7896]
sol.findNumbers(nums)
```

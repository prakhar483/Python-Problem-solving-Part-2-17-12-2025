```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        digits = [int(d) for d in str(n)]
        product = 1
        total_sum = 0
        
        for d in digits:
            product *= d
            total_sum += d
            
        return product - total_sum
```

```python
sol = Solution()

n = 234
sol.subtractProductAndSum(n)
```
